Jelaskan Code dibawah ini (Contoh code untuk pembuktian f(n) = 1+2+3+...+n= n(n+1)/2 dengan menggunakan Induksi Matematika

```
from sympy import *
x, y, z, k, n = symbols('x y z k n')
init_printing(use_unicode=True)

def verify_base_case(formula, valid_from_n, base_case):
    computed_base_case = (sympify(formula)).subs(n, valid_from_n)
    base_case_satisfied = computed_base_case == base_case
    return base_case_satisfied

def expand_induction_hypothesis_k_plus_1(formula, next_expression):
    # inductive hypothesis
    inductive_step = formula.replace('n', 'k') + '+' + next_expression.replace('n', 'k')
    return expand(inductive_step)

def expand_formula_k_plus_1(formula):
    return expand(formula.replace('n', '(k+1)'))

def prove_arithmetic_series_formula(formula, valid_from_n, base_case, next_expression):
    # verify base case
    base_case_satisfied = verify_base_case(formula, valid_from_n, base_case)

    # Inductive step
    inductive_hypothesis = expand_induction_hypothesis_k_plus_1(formula, next_expression)
    formula_next_term = expand_formula_k_plus_1(formula)

    # If the inductive hypothesis matches the formula for the next term
    # then we have proved the formula
    return {
        'Langkah 1 Basis Induksi': base_case_satisfied,
        'Langkah 2 f(k+1)': inductive_hypothesis,
        'Hipotesis Induksi': formula_next_term,
        'Terbukti': base_case_satisfied and inductive_hypothesis == formula_next_term
    }

INPUT_FORMULA = 'n*(n+1)/2'
INPUT_VALID_FROM_N = 1
INPUT_BASE_CASE = 1
INPUT_NEXT_STEP = '(n+1)'

print (prove_arithmetic_series_formula(
    INPUT_FORMULA,
    INPUT_VALID_FROM_N,
    INPUT_BASE_CASE,
    INPUT_NEXT_STEP
    )
)
```

Program di atas merupakan program untuk membuktikan baris dan deret dengan induksi.


---


**Contoh soal:**

Buktikan

$$1+2+3+\ldots+n = \frac{n(n+1)}{2}$$


---


**Pembahasan:**

Untuk membuktikan bahwa persamaan tersebut benar, dapat dilakukan dengan induksi.

Pertama-tama, kita harus mengecek apakah rumus $S(n)$ tersebut berlaku untuk suku pertama. (Cek apakah $S(1) = 1$)

$$S(n) = \frac{n(n+1)}{2}$$

$$S(1) = \frac{1(1+1)}{2}$$

$$S(1) = \frac{1(2)}{2}$$

$$S(1) = \frac{2}{2}$$

$$S(1) = 1$$

Terbukti, rumus $S(n)$ berlaku untuk suku awal.

Kedua, kita asumsikan $n = k$

$$1+2+3+\ldots+n = \frac{n(n+1)}{2}$$

$$1+2+3+\ldots+k = \frac{k(k+1)}{2}$$

Ketiga, cek apakah rumus berlaku untuk $n=k+1$ dengan melibatkan asumsi pada langkah kedua.

$$1+2+3+\ldots+n = \frac{n(n+1)}{2}$$

$$1+2+3+\ldots+(k+1) = \frac{(k+1)((k+1)+1)}{2}$$

$$1+2+3+\ldots+k+(k+1) = \frac{(k+1)(k+2)}{2}$$

Kita ganti $1+2+3+\ldots+k$ dengan asumsi pada tahap kedua

$$\frac{k(k+1)}{2}+(k+1) = \frac{k^2+3k+2}{2}$$

$$\frac{k(k+1)}{2}+\frac{2(k+1)}{2} = \frac{k^2+3k+2}{2}$$

$$k(k+1)+2(k+1) = k^2+3k+2$$

$$k^2+k+2k+2 = k^2+3k+2$$

$$k^2+3k+2 = k^2+3k+2$$

Karena persamaan akhir sama, maka asumsi benar dan rumus $S(n) = \frac{n(n+1)}{2}$ untuk $1+2+3+\ldots+n$ terbukti benar.
