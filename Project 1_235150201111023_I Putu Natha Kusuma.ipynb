```
>>> import sympy
>>> print (sympy.sqrt(7))
>>> import math
>>> print (math.sqrt(7))
```

Apakah perbedaan dari kedua hasil tersebut? Mengapa?

`math.sqrt()` mengembalikan nilai berupa bilangan floating point yang mendekati nilai akar, sedangkan `sympy.sqrt()` mengembalikan nilai asli dari akar tersebut. Jika akar tersebut tidak dapat disederhanakan menjadi bilangan bulat, maka nilai yang dikembalikan oleh `sympy.sqrt()` tetap dalam simbol akar, atau bahkan menyederhanakannya ke bentuk `int(x)*sqrt(y)`, kecuali kita memasukkan `false` ke parameter kedua.

Apa yang dapat Anda simpulkan dari dua hasil kode di atas?

`sympy.Float(num, dps, precision)` mengubah `num` menjadi bentuk floating point. Tak hanya itu, constructor ini juga dapat mengambil parameter `dps` yang akan menentukan berapa angka di belakang koma yang akan dikeluarkan. Jika nilai `dps` lebih sedikit daripada jumlah angka di belakang koma pada `num`, maka akan terjadi pembulatan supaya dapat direpresentasikan dalam `dps` angka di belakang koma.

Catatan: `dps` dan `precision` merupakan parameter tidak wajib, secara default akan bernilai `None` jika tidak dimasukkan nilai.

```
>>> Float(1.6666666,2)
```

```
>>> Float(2.33333,2)
```

Apa yang dapat Anda simpulkan dari hasil dua kode di atas?

Pada kasus dimana angka yang dimasukkan pada `num` berupa floating point, dan angka pada `dps` lebih sedikit dari jumlah digit angka `num`, maka `sympy.Float(num, dps, precision)` akan membulatkan `num` ke nilai bulat terdekat.

Apa yang dapat Anda simpulkan dari ketiga hasil kode di atas?

**Floating Point Number Precision Problem**

Sebagai manusia, kita tentu menggunakan bilangan desimal untuk merepresentasikan angka. Dengan bilangan desimal, kadang kita menemukan masalah saat mengonversikan bilangan pecahan ke floating point, contohnya 1/3. Kita tentu sering menulisnya 0.3333333333, namun sebenarnya nilai floating point dari pecahan tersebut tidak akan pernah bisa kita tuliskan secara akurat. Jadi, dapat kita simpulkan bahawa ada beberapa bilangan yang tidak dapat direpresentasikan dengan bentuk yang akurat dalam bilangan desimal.

Hal serupa juga terjadi pada komputer. Dalam konversi bilangan floating point desimal ke biner, ada beberapa bilangan yang tidak dapat direpresentasikan secara akurat (contohnya 0.4), dan ada pula bilangan yang dapat direpresentasikan dengan akurat (contohnya 0.5).

Untuk itu, kita perlu mengakalinya dengan memberikan `limit_denominator(max_denominator)` dimana `max_denominator` adalah penyebut maksimum dari pecahan yang dihasilkan. Dengan ini, jika terjadi ketidakakuratan pada bilangan floating point, maka akan dibulatkan ke bilangan terdekat dengan penyebut maksimal `max_denominator`.

Catatan: `max_denominator` adalah parameter tidak wajib, jiak tidak diisi maka akan bernilai default (1000000)

Apa yang dapat Anda simpulkan dari kode di atas?

`sympy.S` adalah Singleton Registry. Ini adalah salah satu implementasi dari *Object Oriented Programming*. Singleton yang terdapat pada `sympy.S` adalah class dengan data yang kita sering pakai. Sebagai contoh, `S.One` bernilai sama dengan `Integer(1)`. Namun, jika kita menggunakan `Integer(1)` berkali-kali, ini akan membuat instance berbeda dengan nilai sama berulang kali. Sedangkan dengan menggunakan singleton, kita bisa mengirit memori karena instance yang dibuat hanya satu. Hal ini juga dapat membuat operasi perbandingan berjalan sedikit lebih cepat.

`ask(proposition)` mengevaluasi `proposition` dengan nilai `True` atau `False` jika nilai kebenaran bisa ditentukan. Jika tidak, akan mengembalikan nilai `None`.

Instance `Q` merujuk pada class `AssumptionKeys` yang berisi semua keys yang didukung oleh function `ask(proposition)`.

`ask(Q.complex(x))` mengecek apakah `x` merupakan bilangan kompleks.

`ask(Q.composite(x))` mengecek apakah `x` merupakan bilangan komposit.

`ask(Q.imaginary(x))` mengecek apakah `x` merupakan bilangan imajiner.

`ask(Q.even(x))` mengecek apakah `x` merupakan bilangan genap.

`ask(Q.odd(x))` mengecek apakah `x` merupakan bilangan ganjil.

`ask(Q.posiive(x))` mengecek apakah `x` merupakan bilangan positif.

`ask(Q.negative(x))` mengecek apakah `x` merupakan bilangan negatif.

`ask(Q.zero(x))` mengecek apakah `x` merupakan bilangan nol.

`ask(Q.nonzero(x))` mengecek apakah `x` merupakan bilangan bukan nol.

`ask(Q.rational(x))` mengecek apakah `x` merupakan bilangan rasional.

`ask(Q.irrational(x))` mengecek apakah `x` merupakan bilangan irasional.

```
print("Integer() class represents number to the form of integer and round it towards zero if needed. For example:")


def explain(num, origin=None):
  if origin is None:
    origin = num
  print(str(origin)+" becomes "+str(Integer(num)))


explain(9.89999)
explain(22/7, "22/7")
explain(3/4, "3/4")
```

```
print("S class points towards SingletonRegistry class. It consists of variables that we often use.\n"
"We can use this instead of making a lot of variables with the same value, which is a waste of memory and performance.\n")

def explain(s_str, s, val):
  answer = s.__eq__(val)
  print("Does "+s_str+" equal to "+str(val)+"? "+str(answer))

explain("S.Zero", S.Zero, 0)
explain("S.One", S.One, 1)
explain("S.NaN", S.NaN, nan)
explain("S.Infinity", S.Infinity, oo)
explain("S.ImaginaryUnit", S.ImaginaryUnit, I)
```
