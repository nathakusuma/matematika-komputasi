# Gabungan Karakter

```python
plt.figure(figsize=(10,2.5))

currentX = 0
def plot(list_x, list_y, put_space_after=True):
    global currentX
    list_x = [x + currentX for x in list_x]
    plt.plot(list_x, list_y)
    if put_space_after:
        currentX += 8
    return Matrix([list_x, list_y])

N = plot(Nx,Ny)

A11 = plot(A1x, A1y, False)
A12 = plot(A2x, A2y)

T = plot(Tx,Ty)

H = plot(Hx, Hy)

A21 = plot(A1x, A1y, False)
A22 = plot(A2x, A2y)

K = plot(Kx, Ky)
```

# Jenis Transformasi

```python
def transform(original_matrix, transformation_matrix):
    result_matrix = transformation_matrix * original_matrix
    plt.plot(list(result_matrix[0,:]), list(result_matrix[1,:]))
```

## Identitas

```python
transformation_matrix = Matrix([[1, 0], [0, 1]])
transformation_matrix
```

```python
plt.figure(figsize=(10,2.5))
transform(N, transformation_matrix)
transform(A11, transformation_matrix)
transform(A12, transformation_matrix)
transform(T, transformation_matrix)
transform(H, transformation_matrix)
transform(A21, transformation_matrix)
transform(A22, transformation_matrix)
transform(K, transformation_matrix)
```

## Dilatasi dengan pusat $O (0,0)$

```python
from sympy import Symbol

k = Symbol('k')
```

```python
transformation_matrix = Matrix([[k, 0], [0, k]])
transformation_matrix
```

Misal $k = 5$

```python
transformation_matrix = transformation_matrix.subs({k:5})
transformation_matrix
```

```python
plt.figure(figsize=(10,2.5))
transform(N, transformation_matrix)
transform(A11, transformation_matrix)
transform(A12, transformation_matrix)
transform(T, transformation_matrix)
transform(H, transformation_matrix)
transform(A21, transformation_matrix)
transform(A22, transformation_matrix)
transform(K, transformation_matrix)
```

Tulisan "NATHAK" dilatasi sebanyak 5 kali, dapat dilihat pada nilai absis dan ordinat yang lebih besar jika dibandingkan dengan tulisan yang asli.

## Rotasi sebesar $\theta$ mengelilingi

```python
from sympy import sin, cos

theta = Symbol('')
```

```python
transformation_matrix = Matrix([[cos(theta), -sin(theta)], [sin(theta), cos(theta)]])
transformation_matrix
```

Misal $\theta = 45\degree$

```python
transformation_matrix = transformation_matrix.subs({theta:45})
transformation_matrix
```

```python
plt.figure(figsize=(5,5))
transform(N, transformation_matrix)
transform(A11, transformation_matrix)
transform(A12, transformation_matrix)
transform(T, transformation_matrix)
transform(H, transformation_matrix)
transform(A21, transformation_matrix)
transform(A22, transformation_matrix)
transform(K, transformation_matrix)
```
